```python
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize= (20,10))
plt.plot(df['State/ Union Teritorries'], df['Salary Expenditure 2007-08-Actual'] ) # Remane the column for better understanding

plt.show()
```

```python
# drop total related column : Sub Total - SCS, Sub Total - GCS,Sub Total UT.s from State/ Union Teritorries
nottotal = df.drop([df.index[33], df.index[29], df.index[32], df.index[11]])
```

```python
nottotal.head(35)
```

```python
# Visualize via line graph
plt.figure(figsize= (20,10))
plt.plot(nottotal['State/ Union Teritorries'], nottotal['Salary Expenditure 2007-08-Actual'] )

plt.show()
```

```python
# Visualize via bar graph
fig, ax = plt.subplots(figsize =(20,10)) #plt.figure(figsize= (20,10))

plt.bar(nottotal['State/ Union Teritorries'], nottotal['Salary Expenditure 2007-08-Actual'])

fig.text(0.9, 0.15, 'Jeeteshgavande30', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
plt.show()
```

```python
# Visualize via pie graph
plt.figure(figsize= (20,12))
count= nottotal['State/ Union Teritorries']
plt.pie(nottotal['Salary Expenditure 2007-08-Actual'],labels= count, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
